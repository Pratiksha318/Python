```python
import matplotlib.pyplot as plt

# pyplot is a collection of function
```

```python
#Create chart
X=[1,2,3,4,5]
Y=[10,20,25,35,45]
plt.plot(X,Y)   # Line chart
plt.show()      # Display chart in jp notebook
```

```python
# Customize chart
plt.figure(figsize=(5,4))  
plt.plot(X, Y, color='red', marker='o', linestyle='dashed',linewidth=2, markersize=12)

plt.title("Line Chart")
plt.xlabel('X-axis') 
plt.ylabel('Y-axis')

plt.show()
```

```python
# Advance line chart
X=[1,2,3,4,5]
Y=[10,20,25,35,45]
Y1=[20,30,35,45,55]
plt.plot(X,Y,label="Sales 2024")  
plt.plot(X,Y1,label="Sales 2025")
plt.title('YOY Sales')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
# Bar chart
X=["A","B","C","D","E"]
Y=[15,20,55,35,45]
plt.bar(X,Y)   # Bar chart
plt.title("Bar chart")
plt.show()
```

```python
# Histogram  continous data
import random
data= [random.randint(1,10) for _ in range(100)]
plt.hist(data,bins=10)
plt.title("Histogram")
plt.show()
```

```python
# Pie chart
Category=["A","B","C","D","E"]
Sales=[15,20,55,35,45]
plt.pie(Sales,labels=Category,autopct="%1.1f%%")
plt.title("Pie chart")
plt.show()
    
```

```python
# Scatter chart  used to find relationship between variables
y1=[10,20,25,35,45]
y2=[20,30,35,45,55]
plt.scatter(y1,y2)
plt.title("Scatter Plot")
plt.show
```

```python
# Subplots   used to show multiple charts in one figure
Category=["Mon","Tue","Wed","Thu","Fri"]
Sales=[15,20,55,35,45]

y1=[10,20,25,35,45]
y2=[20,30,35,45,55]

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)  # row,column,position
plt.bar(Category,Sales)
plt.title("Daily sales")


plt.subplot(1,2,2)
plt.scatter(y1,y2)
plt.title("Scatter plot")

plt.show()
```
